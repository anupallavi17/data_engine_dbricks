### 49. Write a Python script to display the

- c. Month of year
- d. Week number of the year

- f. Day of year
- g. Day of the month
- h. Day of week

```python
import datetime as dt
current_date_time = dt.datetime.now()
print("Current date and time is:",current_date_time)
print("Current year is ",current_date_time.year )
print("Month of year is",current_date_time.month)
print("Week Number of the year is",(dt.datetime.strftime(current_date_time, '%U')))
print("weekday of week is",current_date_time.weekday())
print("Day of year is",current_date_time.strftime('%j'))
print("Day of month is",current_date_time.day)
print("Day of week is ",current_date_time.strftime('%A'))
```
