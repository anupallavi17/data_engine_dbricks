#### 1. Merge the datasets Customers, Product Hierarchy and Transactions as Customer_Final. Ensure to keep all customers who have done transactions with us and select the join type accordingly.

```python
import pandas as pd
import numpy as np
import datetime as dt
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Customers = pd.read_csv('Customer.csv')
Product = pd.read_csv('prod_cat_info.csv')
Transactions = pd.read_csv('Transactions.csv')
```

```python
Custtran = pd.merge(Customers,Transactions,left_on='customer_Id', right_on='cust_id',how="right")
```

```python
#below line shows that customer_Id and cust_id are same
np.where(Custtran.customer_Id != Custtran.cust_id)
```

```python
# dropping cust_id column 
Custtran.drop('cust_id',inplace=True,axis=1)
Custtran
```

```python
Customer_Final=pd.merge(Custtran,Product,left_on=['prod_cat_code','prod_subcat_code'],right_on=['prod_cat_code','prod_sub_cat_code'],how='left')
Customer_Final
```

```python
np.where(Customer_Final.prod_sub_cat_code != Customer_Final.prod_subcat_code)
```

```python
Customer_Final.drop('prod_sub_cat_code',inplace=True,axis=1)
Customer_Final
```

### 5. Analyze which product categories are more popular among females vs male customers

```python
prod_cat_sort = Customer_Final.groupby(['Gender','prod_cat'])['prod_cat'].count().sort_values(ascending =False)
prod_pop_male = prod_cat_sort.loc['M'].head(1)
prod_pop_Female = prod_cat_sort.loc['F'].head(1)
print ("Popular product among Female: ",prod_pop_Female)
print ("Popular product among Male: ",prod_pop_male)
```

#### 8.What was the total amount earned from the "Electronics" and "Clothing" categories from Flagship Stores?

```python
Customer_Final.loc[(Customer_Final.prod_cat=='Electronics') | (Customer_Final.prod_cat == 'Clothing'),:].where(Customer_Final.Store_type=='Flagship store')['total_amt'].sum()
```

### 9. What was the total amount earned from "Male" customers under the "Electronics" category?

```python
Customer_Final.loc[(Customer_Final.Gender == 'M') & (Customer_Final.prod_cat == 'Electronics'),:]['total_amt'].sum()
```

### 10. How many customers have more than 10 unique transactions, after removing all transactions which have any negative amounts?

```python
# to Remove the negative amounts
pre_index = pd.DataFrame(Customer_Final.loc[Customer_Final.total_amt < 0,:]).index
Customer_Final.drop(index= pre_index , inplace =True)
# to display customers with more than 10 unique transactions
cust_unique_tran= Customer_Final.groupby('customer_Id')['transaction_id'].nunique().sort_values(ascending=False).reset_index()
cust_unique_tran.loc[cust_unique_tran.transaction_id > 10,:]
```

### 11. For all customers aged between 25 - 35, find out:

```python
Customer_Final.DOB = pd.to_datetime(Customer_Final.DOB,format='%d-%m-%Y')
current_year = dt.datetime.now().year
Customer_Final['age'] = current_year - Customer_Final.DOB.dt.year
Customer_summary = Customer_Final.loc[(Customer_Final['age'] > 25) & (Customer_Final['age'] < 35) , :]
Customer_summary
```

#### a. What was the total amount spent for “Electronics” and “Books” product categories?

```python
Customer_summary.loc[(Customer_summary.prod_cat == 'Electronics') | (Customer_summary.prod_cat == 'Books')]['total_amt'].sum()
```

#### b. What was the total amount spent by these customers between 1st Jan, 2014 to 1st Mar, 2014

```python
lower_date = pd.to_datetime('1-jan-2014',format ='%d-%b-%Y')
upper_date = pd.to_datetime('1-Mar-2014',format ='%d-%b-%Y')
Customer_summary['tran_date']=pd.to_datetime(Customer_summary.tran_date,format='mixed')
Customer_summary.loc[(Customer_summary.tran_date > lower_date) & (Customer_summary.tran_date < upper_date),['total_amt']].sum()
```
